```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Generate variabel universe
#   * Kualitas dan pelayanan pada rentang subjektif [0, 10]
#   * Kepuasan pada rentang [0, 10]
x_qual = np.arange(0, 11, 1)
x_serv = np.arange(0, 11, 1)
x_satisf = np.arange(0, 11, 1)

# Generate fungsi keanggotaan fuzzy
qual_lo = fuzz.trimf(x_qual, [0, 0, 5])
qual_md = fuzz.trimf(x_qual, [0, 5, 10])
qual_hi = fuzz.trimf(x_qual, [5, 10, 10])
serv_lo = fuzz.trimf(x_serv, [0, 0, 5])
serv_md = fuzz.trimf(x_serv, [0, 5, 10])
serv_hi = fuzz.trimf(x_serv, [5, 10, 10])
satisf_lo = fuzz.trimf(x_satisf, [0, 0, 5])
satisf_md = fuzz.trimf(x_satisf, [0, 5, 10])
satisf_hi = fuzz.trimf(x_satisf, [5, 10, 10])

# Visualisasikan variabel universe dan fungsi keanggotaan
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(8, 9))

ax0.plot(x_qual, qual_lo, 'b', linewidth=1.5, label='Buruk')
ax0.plot(x_qual, qual_md, 'g', linewidth=1.5, label='Biasa')
ax0.plot(x_qual, qual_hi, 'r', linewidth=1.5, label='Baik')
ax0.set_title('Kualitas Produk')
ax0.legend()

ax1.plot(x_serv, serv_lo, 'b', linewidth=1.5, label='Buruk')
ax1.plot(x_serv, serv_md, 'g', linewidth=1.5, label='Biasa')
ax1.plot(x_serv, serv_hi, 'r', linewidth=1.5, label='Baik')
ax1.set_title('Pelayanan Customer')
ax1.legend()

ax2.plot(x_satisf, satisf_lo, 'b', linewidth=1.5, label='Rendah')
ax2.plot(x_satisf, satisf_md, 'g', linewidth=1.5, label='Menengah')
ax2.plot(x_satisf, satisf_hi, 'r', linewidth=1.5, label='Tinggi')
ax2.set_title('Kepuasan Customer')
ax2.legend()

# Matikan sumbu atas/kanan
for ax in (ax0, ax1, ax2):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()

```

```python
# Asumsi x_qual, qual_lo, qual_md, qual_hi, x_serv, serv_lo, serv_md, serv_hi, x_satisf, satisf_lo, satisf_md, satisf_hi sudah didefinisikan

# Nilai input
qual_input = 8
serv_input = 6

# Derajat keanggotaan untuk nilai input
qual_level_lo = fuzz.interp_membership(x_qual, qual_lo, qual_input)
qual_level_md = fuzz.interp_membership(x_qual, qual_md, qual_input)
qual_level_hi = fuzz.interp_membership(x_qual, qual_hi, qual_input)

serv_level_lo = fuzz.interp_membership(x_serv, serv_lo, serv_input)
serv_level_md = fuzz.interp_membership(x_serv, serv_md, serv_input)
serv_level_hi = fuzz.interp_membership(x_serv, serv_hi, serv_input)

# Aturan 1: Jika kualitas rendah DAN pelayanan rendah, maka kepuasan rendah
rule1 = np.fmin(qual_level_lo, serv_level_lo)
satisf_activation_lo1 = np.fmin(rule1, satisf_lo)

# Aturan 2: Jika kualitas rendah DAN pelayanan sedang, maka kepuasan rendah
rule2 = np.fmin(qual_level_lo, serv_level_md)
satisf_activation_lo2 = np.fmin(rule2, satisf_lo)

# Aturan 3: Jika kualitas rendah DAN pelayanan tinggi, maka kepuasan menengah
rule3 = np.fmin(qual_level_lo, serv_level_hi)
satisf_activation_md1 = np.fmin(rule3, satisf_md)

# Aturan 4: Jika kualitas sedang DAN pelayanan rendah, maka kepuasan rendah
rule4 = np.fmin(qual_level_md, serv_level_lo)
satisf_activation_lo3 = np.fmin(rule4, satisf_lo)

# Aturan 5: Jika kualitas sedang DAN pelayanan sedang, maka kepuasan menengah
rule5 = np.fmin(qual_level_md, serv_level_md)
satisf_activation_md2 = np.fmin(rule5, satisf_md)

# Aturan 6: Jika kualitas sedang DAN pelayanan tinggi, maka kepuasan tinggi
rule6 = np.fmin(qual_level_md, serv_level_hi)
satisf_activation_hi1 = np.fmin(rule6, satisf_hi)

# Aturan 7: Jika kualitas tinggi DAN pelayanan rendah, maka kepuasan menengah
rule7 = np.fmin(qual_level_hi, serv_level_lo)
satisf_activation_md3 = np.fmin(rule7, satisf_md)

# Aturan 8: Jika kualitas tinggi DAN pelayanan sedang, maka kepuasan tinggi
rule8 = np.fmin(qual_level_hi, serv_level_md)
satisf_activation_hi2 = np.fmin(rule8, satisf_hi)

# Aturan 9: Jika kualitas tinggi DAN pelayanan tinggi, maka kepuasan tinggi
rule9 = np.fmin(qual_level_hi, serv_level_hi)
satisf_activation_hi3 = np.fmin(rule9, satisf_hi)

# Max dari masing masing nilai linguistik dari rule
satisf_activation_lo = np.fmax(satisf_activation_lo1,
                              np.fmax(satisf_activation_lo2, satisf_activation_lo3))
satisf_activation_md = np.fmax(satisf_activation_md1,
                              np.fmax(satisf_activation_md2, satisf_activation_md3))
satisf_activation_hi = np.fmax(satisf_activation_hi1,
                              np.fmax(satisf_activation_hi2, satisf_activation_hi3))

# Visualisasikan ini
fig, ax0 = plt.subplots(figsize=(8, 3))

ax0.fill_between(x_satisf, np.zeros_like(x_satisf), satisf_activation_lo, facecolor='b', alpha=0.7)
ax0.plot(x_satisf, satisf_lo, 'b', linewidth=0.5, linestyle='--')

ax0.fill_between(x_satisf, np.zeros_like(x_satisf), satisf_activation_md, facecolor='g', alpha=0.7)
ax0.plot(x_satisf, satisf_md, 'g', linewidth=0.5, linestyle='--')

ax0.fill_between(x_satisf, np.zeros_like(x_satisf), satisf_activation_hi, facecolor='r', alpha=0.7)
ax0.plot(x_satisf, satisf_hi, 'r', linewidth=0.5, linestyle='--')

ax0.set_title('Aktivitas keanggotaan output')

# Matikan sumbu atas/kanan
for ax in (ax0,):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()

# Agregasikan semua fungsi keanggotaan output
aggregated = np.fmax(satisf_activation_lo,
                     np.fmax(satisf_activation_md, satisf_activation_hi))

# Defuzzifikasi hasil
satisfaction = fuzz.defuzz(x_satisf, aggregated, 'centroid')
satisfaction_activation = fuzz.interp_membership(x_satisf, aggregated, satisfaction)

# Visualisasikan output akhir
fig, ax1 = plt.subplots(figsize=(8, 3))

ax1.plot(x_satisf, satisf_lo, 'b', linewidth=0.5, linestyle='--')
ax1.plot(x_satisf, satisf_md, 'g', linewidth=0.5, linestyle='--')
ax1.plot(x_satisf, satisf_hi, 'r', linewidth=0.5, linestyle='--')
ax1.fill_between(x_satisf, np.zeros_like(x_satisf), aggregated, facecolor='Orange', alpha=0.7)
ax1.plot([satisfaction, satisfaction], [0, satisfaction_activation], 'k', linewidth=1.5, alpha=0.9)
ax1.set_title('Agregasi keanggotaan dan hasil (garis)')

# Matikan sumbu atas/kanan
for ax in (ax1,):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
plt.show()

print(f'Tingkat Kepuasan: {satisfaction:f}')
print(f'Derajat Keanggotaan Kepuasan: {satisfaction_activation:f}')

```
